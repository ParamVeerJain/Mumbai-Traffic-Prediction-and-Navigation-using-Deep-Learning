```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
sns.set(style="darkgrid")
CHUNK_SIZE = 200_000      
```

```python
FILE_PATH='D:/Urban-Traffic-and-Parking-Analysis-using-LSTM-Autoencoders-and-Reinforcement-Learning/Dataset/big_data.csv'
```

```python
num_cols = [
    'length_m', 'free_flow_speed', 'congestion',
    'travel_time_sec', 'rain', 'incident', 'speed_ratio'
]

sum_vals = defaultdict(float)
sum_sq_vals = defaultdict(float)
count_vals = defaultdict(int)
min_vals = defaultdict(lambda: np.inf)
max_vals = defaultdict(lambda: -np.inf)

for chunk in pd.read_csv(FILE_PATH, chunksize=CHUNK_SIZE):

    for col in num_cols:
        data = chunk[col].dropna()

        sum_vals[col] += data.sum()
        sum_sq_vals[col] += (data ** 2).sum()
        count_vals[col] += len(data)

        min_vals[col] = min(min_vals[col], data.min())
        max_vals[col] = max(max_vals[col], data.max())

stats = {}

for col in num_cols:
    mean = sum_vals[col] / count_vals[col]
    var = (sum_sq_vals[col] / count_vals[col]) - (mean ** 2)
    std = np.sqrt(var)

    stats[col] = {
        "mean": mean,
        "std": std,
        "min": min_vals[col],
        "max": max_vals[col]
    }

pd.DataFrame(stats).T
```
