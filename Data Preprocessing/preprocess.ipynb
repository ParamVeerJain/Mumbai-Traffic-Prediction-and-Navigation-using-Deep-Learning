```python
dataFrame['length_log']=np.log1p(dataFrame['length_m'])
dataFrame["speed_ratio"] = (dataFrame["current_speed"] / dataFrame["free_flow_speed"]).clip(0.0, 1.0)
dataFrame.head()
```

```python
dataFrame=dataFrame.drop(['travel_time_sec','u_lat','v_lat','v_lon','u_lon'],axis=1)
dataFrame.head()
```

```python
import os

OUTPUT_FILE = "lstm_dataset_1.csv"
BATCH_SIZE = 50_000  # tune this (50k–500k depending on RAM)

if os.path.exists(OUTPUT_FILE):
    os.remove(OUTPUT_FILE)

def save_in_batches(dataFrame):
    total_rows = len(dataFrame)
    
    for start in range(0, total_rows, BATCH_SIZE):
        end = start + BATCH_SIZE
        batch = dataFrame.iloc[start:end]

        # First batch → write normally
        if start == 0:
            batch.to_csv(OUTPUT_FILE, index=False, mode='w')
        else:
            batch.to_csv(OUTPUT_FILE, index=False, mode='a', header=False)

        print(f"Saved rows {start} → {end}")

# Example usage
# df = your 24M dataframe
save_in_batches(dataFrame)
```
